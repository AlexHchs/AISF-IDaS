```python
train_test_overview = {}
for attack_type in np.unique(y_malicious_test):
    for attack_impl, count in attack_type_train[y_malicious_train == attack_type].value_counts().items():
        if (attack_type, attack_impl) not in train_test_overview:
            train_test_overview[(attack_type, attack_impl)] = {}
        train_test_overview[(attack_type, attack_impl)]['train'] = count
    for attack_impl, count in attack_type_test[y_malicious_test == attack_type].value_counts().items():
        if (attack_type, attack_impl) not in train_test_overview:
            train_test_overview[(attack_type, attack_impl)] = {}
        train_test_overview[(attack_type, attack_impl)]['test'] = count
pd.DataFrame.from_dict(train_test_overview, orient="index").rename_axis(["Class", "Impl"])
```
